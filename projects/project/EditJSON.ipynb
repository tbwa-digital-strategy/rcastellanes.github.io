```python
import pandas as pd
import json
import geopandas as gpd
```

```python
county = gpd.read_file('county.json')
```

```python
county.head()
```

```python
maindata = pd.read_csv('newmaindata.csv')
```

```python
maindata.head()
```

```python
county.STATEFP = pd.Series(county.STATEFP, dtype = 'float64')
county.COUNTYFP = pd.Series(county.COUNTYFP, dtype = 'float64')
```

```python
subdata = pd.read_csv('newsubdata.csv')
```

```python
subdata.head()
```

```python
newdata = pd.merge(county, maindata, how = 'left', left_on = ['STATEFP', 'COUNTYFP'], right_on=['STATE','COUNTY'])
newdata = pd.merge(newdata, subdata, how = 'left', left_on = ['STATEFP', 'COUNTYFP'], right_on=['STATE','COUNTY'])
```

```python
newdata.to_file('countymerged.json', driver = 'GeoJSON')
```
